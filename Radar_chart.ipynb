```python
import pandas as pd
import numpy as np
import random

#Creem un dataframe amb valors aleatòris
np.random.seed(42)
characters = ['Personatge 1', 'Personatge 2']
stats = ['Força', 'Intel·ligència', 'Agilitat', 'Resistència', 'Carisma', 'Sort', 'Defensa', 'Vida']

data = {}
for character in characters:
    data[character] = [random.randint(3, 8) for _ in range(len(stats))]

df = pd.DataFrame(data, index=stats)


print(df)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6), subplot_kw=dict(polar=True))

angles = np.linspace(0, 2*np.pi, len(stats), endpoint=False).tolist()
angles += angles[:1]

for character in characters:
    values = df[character].values.tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=2, label=character)
    ax.fill(angles, values, alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(stats)
ax.set_yticks(range(0, 11, 2))
ax.set_yticklabels(range(0, 11, 2))
ax.set_ylim(0, 10)
plt.title('Comparació de personatges')
plt.legend()

plt.tight_layout()
plt.show()
```
